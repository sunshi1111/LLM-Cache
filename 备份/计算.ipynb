```python
def calculate_decoder_stats(input_text):
    """
    计算decoder层和decoder其余层的总耗时
    
    参数:
        input_text (str): 包含decoder层和decoder其余层耗时信息的文本
        
    返回:
        tuple: (decoder层总耗时, decoder其余层总耗时, decoder层记录数, decoder其余层记录数)
    """
    # 分割文本为行
    lines = input_text.strip().split('\n')
    
    decoder_total_time = 0.0
    decoder_rest_total_time = 0.0
    decoder_count = 0
    decoder_rest_count = 0
    
    for line in lines:
        line = line.strip()
        if not line:  # 跳过空行
            continue
            
        # 检查行是否包含decoder层耗时信息
        if "decoder层耗时" in line:
            try:
                # 提取时间值
                time_str = line.split(":")[-1].strip()
                # 移除"秒"并转换为浮点数
                if "秒" in time_str:
                    time_value = float(time_str.replace("秒", ""))
                else:
                    time_value = float(time_str)
                
                decoder_total_time += time_value
                decoder_count += 1
            except Exception as e:
                print(f"解析行 '{line}' 时出错: {e}")
                
        # 检查行是否包含decoder其余层耗时信息
        elif "decoder其余层耗时" in line:
            try:
                # 提取时间值
                time_str = line.split(":")[-1].strip()
                # 移除"秒"并转换为浮点数
                if "秒" in time_str:
                    time_value = float(time_str.replace("秒", ""))
                else:
                    time_value = float(time_str)
                
                decoder_rest_total_time += time_value
                decoder_rest_count += 1
            except Exception as e:
                print(f"解析行 '{line}' 时出错: {e}")
                
    return decoder_total_time, decoder_rest_total_time, decoder_count, decoder_rest_count

# 输入数据
input_data = """
decoder层耗时: 0.000471秒
本次第二次循环耗时: 0.080239秒
累计第二次循环总耗时: 0.080239秒
decoder其余层耗时: 0.124124秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083180秒
累计第二次循环总耗时: 0.163419秒
decoder其余层耗时: 0.123339秒
decoder层耗时: 0.000471秒
本次第二次循环耗时: 0.082552秒
累计第二次循环总耗时: 0.245972秒
decoder其余层耗时: 0.123262秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.083271秒
累计第二次循环总耗时: 0.329243秒
decoder其余层耗时: 0.124695秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.082770秒
累计第二次循环总耗时: 0.412013秒
decoder其余层耗时: 0.124585秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.082681秒
累计第二次循环总耗时: 0.494694秒
decoder其余层耗时: 0.124776秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.082898秒
累计第二次循环总耗时: 0.577592秒
decoder其余层耗时: 0.125225秒
decoder层耗时: 0.000477秒
本次第二次循环耗时: 0.083655秒
累计第二次循环总耗时: 0.661247秒
decoder其余层耗时: 0.124507秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.083107秒
累计第二次循环总耗时: 0.744353秒
decoder其余层耗时: 0.124558秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.082960秒
累计第二次循环总耗时: 0.827314秒
decoder其余层耗时: 0.123862秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.083309秒
累计第二次循环总耗时: 0.910622秒
decoder其余层耗时: 0.123799秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.082446秒
累计第二次循环总耗时: 0.993068秒
decoder其余层耗时: 0.124001秒
decoder层耗时: 0.000470秒
本次第二次循环耗时: 0.082574秒
累计第二次循环总耗时: 1.075642秒
decoder其余层耗时: 0.124650秒
decoder层耗时: 0.000468秒
本次第二次循环耗时: 0.083390秒
累计第二次循环总耗时: 1.159032秒
decoder其余层耗时: 0.124713秒
decoder层耗时: 0.000477秒
本次第二次循环耗时: 0.083209秒
累计第二次循环总耗时: 1.242241秒
decoder其余层耗时: 0.124474秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.082809秒
累计第二次循环总耗时: 1.325050秒
decoder其余层耗时: 0.124662秒
decoder层耗时: 0.000471秒
本次第二次循环耗时: 0.083371秒
累计第二次循环总耗时: 1.408421秒
decoder其余层耗时: 0.123634秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.082935秒
累计第二次循环总耗时: 1.491357秒
decoder其余层耗时: 0.123869秒
decoder层耗时: 0.000470秒
本次第二次循环耗时: 0.082604秒
累计第二次循环总耗时: 1.573960秒
decoder其余层耗时: 0.124576秒
decoder层耗时: 0.000492秒
本次第二次循环耗时: 0.083272秒
累计第二次循环总耗时: 1.657233秒
decoder其余层耗时: 0.124785秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083272秒
累计第二次循环总耗时: 1.740505秒
decoder其余层耗时: 0.124832秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.083272秒
累计第二次循环总耗时: 1.823776秒
decoder其余层耗时: 0.123684秒
decoder层耗时: 0.000472秒
本次第二次循环耗时: 0.082695秒
累计第二次循环总耗时: 1.906471秒
decoder其余层耗时: 0.124779秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083225秒
累计第二次循环总耗时: 1.989696秒
decoder其余层耗时: 0.124020秒
decoder层耗时: 0.000476秒
本次第二次循环耗时: 0.083049秒
累计第二次循环总耗时: 2.072746秒
decoder其余层耗时: 0.125459秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083981秒
累计第二次循环总耗时: 2.156727秒
decoder其余层耗时: 0.124757秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.083298秒
累计第二次循环总耗时: 2.240025秒
decoder其余层耗时: 0.124573秒
decoder层耗时: 0.000476秒
本次第二次循环耗时: 0.083317秒
累计第二次循环总耗时: 2.323343秒
decoder其余层耗时: 0.123832秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083307秒
累计第二次循环总耗时: 2.406650秒
decoder其余层耗时: 0.123803秒
decoder层耗时: 0.000476秒
本次第二次循环耗时: 0.083278秒
累计第二次循环总耗时: 2.489928秒
decoder其余层耗时: 0.124155秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.083421秒
累计第二次循环总耗时: 2.573349秒
decoder其余层耗时: 0.124454秒
decoder层耗时: 0.000476秒
本次第二次循环耗时: 0.083357秒
累计第二次循环总耗时: 2.656706秒
decoder其余层耗时: 0.124705秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083377秒
累计第二次循环总耗时: 2.740082秒
decoder其余层耗时: 0.124501秒
decoder层耗时: 0.000478秒
本次第二次循环耗时: 0.083430秒
累计第二次循环总耗时: 2.823513秒
decoder其余层耗时: 0.124743秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083414秒
累计第二次循环总耗时: 2.906927秒
decoder其余层耗时: 0.124680秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083252秒
累计第二次循环总耗时: 2.990179秒
decoder其余层耗时: 0.124220秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083400秒
累计第二次循环总耗时: 3.073579秒
decoder其余层耗时: 0.124442秒
decoder层耗时: 0.000473秒
本次第二次循环耗时: 0.083341秒
累计第二次循环总耗时: 3.156920秒
decoder其余层耗时: 0.124724秒
decoder层耗时: 0.000478秒
本次第二次循环耗时: 0.083330秒
累计第二次循环总耗时: 3.240250秒
decoder其余层耗时: 0.125742秒
decoder层耗时: 0.000477秒
本次第二次循环耗时: 0.083837秒
累计第二次循环总耗时: 3.324087秒
decoder其余层耗时: 0.124817秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083270秒
累计第二次循环总耗时: 3.407357秒
decoder其余层耗时: 0.125742秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083666秒
累计第二次循环总耗时: 3.491023秒
decoder其余层耗时: 0.124384秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083420秒
累计第二次循环总耗时: 3.574443秒
decoder其余层耗时: 0.124402秒
decoder层耗时: 0.000481秒
本次第二次循环耗时: 0.083481秒
累计第二次循环总耗时: 3.657925秒
decoder其余层耗时: 0.125026秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083416秒
累计第二次循环总耗时: 3.741341秒
decoder其余层耗时: 0.124620秒
decoder层耗时: 0.000475秒
本次第二次循环耗时: 0.083235秒
累计第二次循环总耗时: 3.824576秒
decoder其余层耗时: 0.124630秒
decoder层耗时: 0.000492秒
本次第二次循环耗时: 0.083486秒
累计第二次循环总耗时: 3.908062秒
decoder其余层耗时: 0.124866秒
decoder层耗时: 0.000474秒
本次第二次循环耗时: 0.083240秒
累计第二次循环总耗时: 3.991302秒
decoder其余层耗时: 0.124782秒
model层耗时: 9.996048秒

"""

# 计算decoder层和decoder其余层的总耗时
decoder_time, decoder_rest_time, decoder_count, decoder_rest_count = calculate_decoder_stats(input_data)

# 输出统计结果
print(f"找到 {decoder_count} 条decoder层记录")
print(f"decoder层总耗时: {decoder_time:.6f}秒")
print(f"decoder层平均耗时: {decoder_time/decoder_count:.6f}秒")
print()
print(f"找到 {decoder_rest_count} 条decoder其余层记录")
print(f"decoder其余层总耗时: {decoder_rest_time:.6f}秒")
print(f"decoder其余层平均耗时: {decoder_rest_time/decoder_rest_count:.6f}秒")
print()
print(f"总计算时间(decoder层+其余层): {decoder_time + decoder_rest_time:.6f}秒")
print(f"decoder层占比: {(decoder_time/(decoder_time + decoder_rest_time))*100:.2f}%")
print(f"decoder其余层占比: {(decoder_rest_time/(decoder_time + decoder_rest_time))*100:.2f}%")
```
